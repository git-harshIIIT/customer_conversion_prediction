# **Customer Conversion Prediction**


In an insurance company the employ tries mutiple outreach plans to sell term insurance to the customers.And most common way they try to do so is Telephonic marketing campaignsh, however they have lot of cost. So it is important to identify the customers that are most likely to convert so that they can be targeted via call.


We try to build an ML model that will predict if a client will subscribe to the insurance using the historical marketing data of an insurance company but before that we will do proper analysis of to identify the rank of importance of features.


### **Features**:

*  age (numeric)
*   job : type of job
*   marital : marital status
*   educational_qual : education status
*   call_type : contact communication type
*   day: last contact day of the month (numeric)
*   mon: last contact month of year
*   dur: last contact duration, in seconds (numeric)
*   num_calls: number of contacts performed during this campaign and for this   client
*   prev_outcome: outcome of the previous marketing campaign (categorical: "unknown","other","failure","success")


*   Output variable (desired target):
y - has the client subscribed to the insurance?

### **Outlier Deduction and Correction**
**Outlier Detection**
1.   Z-Score
         Z-Score(x)=(x-mean(x)) / SD(x)
      
      **Threshold Limit**
      Z-Score > 3 and Z-Score < -3 ---> Outlier
2.   IQR (Inter Quartile Range)

          IQR = Q3(75%)-Q1(25%)
      
      **Upper Threshold** = Q3 + (1.5 * IQR)
      
      **Lower Threshold** = Q1 - (1.5 * IQR)
3.   Plotting
      Box Plot

**Outlier Correction**
1.   Deletion
2.   Clip/Strip



```python
df.info()
```

## **Age**

**Box Plot**

```python
sns.set(style="whitegrid")
sns.boxplot(x=df['age'], color='Chartreuse')
```

From outlier we can see that there are many dots are displayed outside vertical lines .

**IQR**

```python
q1,q3=np.percentile(df["age"],[25,75])
IQR=q3-q1
upper=q3+1.5*IQR
lower=q1-1.5*IQR
print("Upper age bound:",upper,"Lower age bound :", lower)
```

**Removing outlier for Age**

```python
df.age = df.age.clip(10.5,70.5)
```

```python
# X --> Feature
# y-- > Target

x = df[['age', 'job', 'marital', 'education_qual', 'call_type', 'day', 'mon', 'dur', 'num_calls', 'prev_outcome']].values
y=df['target'].values
```

# **Spliting**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state = 3 )
```

```python
nan_indices = np.isnan(x_train).any(axis=1)
print("Number of rows with NaN values:", nan_indices.sum())
```

```python
print(x_train[nan_indices])
```

# **Balancing**

```python
#(Synthetic Minority Over-sampling Technique +  Edited Nearest Neighbors) - SMOTeENN
from imblearn.combine import SMOTEENN
smt = SMOTEENN(sampling_strategy='all')
x_train_smt, y_train_smt = smt.fit_resample(x_train, y_train)
```

```python
print(len(x_train_smt))
print(len(y_train_smt))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_smt)
x_test_scaled = scaler.transform(x_test)
```
